# Aprendizaje en entornos complejos 

En esta sección se estudian dos clases de algoritmos utilizados en problemas de aprendizaje complejos, esto es, aquellos en los cuales el estado del problema cambia de un paso temporal al siguiente. En particular, se implementan métodos llamados "tabulares" y métodos "aproximados". Los primeros utilizan matrices en las que almacenan la calidad estimada de cada par (estado, acción), de tal forma que el agente pueda determinar, para cualquier estado en el que se encuentre, cual es la acción que maximiza las recompensas a largo plazo. Sin embargo, en caso de que el espacio de estados sea excesivamente grande o incluso infinito, no es posible almacenar estos valores en una matriz, sino que se precisa de técnicas capaces de parametrizar los estados, ya sea mediante discretización o mediante el uso de funciones no lineales, como es el caso de las redes neuronales.
