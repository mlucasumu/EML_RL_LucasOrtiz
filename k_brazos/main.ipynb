# Problema del bandido de los k brazos

Esta sección del proyecto se centra en analizar tres familias de algoritmos para la resolución del problema del bandido de k brazos: métodos $\epsilon$-greedy, métodos UCB y métodos de ascenso de gradiente. A su vez, cada familia consta de diversos algoritmos, cada uno de los cuales ha sido puesto a prueba con bandidos de diferente naturaleza. En concreto, se han probado bandidos cuyos brazos siguen una distribución normal, otros con brazos de distribución binomial y, por último, brazos que siguen una distribución Bernoulli.

Los tres estudios realizados pueden ser accedidos a través de los siguientes enlaces:

1. **[Métodos $\epsilon$-greedy](./epsilon_greedy.ipynb)**

2. **[Métodos UCB](./UCB_algorithms.ipynb)**

3. **[Métodos de ascenso de gradiente](./ascenso_gradiente.ipynb)**
